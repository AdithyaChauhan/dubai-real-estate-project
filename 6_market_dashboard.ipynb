```sql
%python
storage_account_name = "REMOVED_FOR_SECURITY"

storage_account_key = "REMOVED_FOR_SECURITY"

spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    storage_account_key
)

gold_path = f"abfss://gold@{storage_account_name}.dfs.core.windows.net"

spark.read.format("delta").load(f"{gold_path}/sales_overview").createOrReplaceTempView("gold_sales_overview")
spark.read.format("delta").load(f"{gold_path}/area_stats").createOrReplaceTempView("gold_area_stats")
spark.read.format("delta").load(f"{gold_path}/supply_stats").createOrReplaceTempView("gold_supply_stats")

print("✅ Views ready.")
```

```sql
SELECT month, total_sales_aed FROM gold_sales_overview ORDER BY month
```

```sql
SELECT area_name, total_volume_aed FROM gold_area_stats
ORDER BY total_volume_aed DESC LIMIT 10
```

```sql
SELECT room_type, unit_count FROM gold_supply_stats
```
